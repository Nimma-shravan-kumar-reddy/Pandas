# Filtering Data

```python
df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') ]
df.loc[(df['Type 1']=='Grass') | (df['Type 2']=='Poison')]
```

```python
# filtering out mega pokemons

df.loc[df['Name'].str.contains('Mega')]
```

```python
# No mega pokemons i.e only normal pokemons

df.loc[~df['Name'].str.contains('Mega')]
```

```python
# using regular expresions
import re

df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)] # re.I ingnore case
```

```python
# filtering pokemon data that starts with name pi
df.loc[df['Name'].str.contains('^Pi[a-z]*' , flags=re.I, regex= True)]
```

# Conditional changes

```python
df.loc[df['Type 1']=='Fire', 'Type 1']='Flamer'
df
```

```python
df.loc[df['Type 1']=='Flamer', 'Type 1']='Fire'
df
```

# Aggregate statistics (Groupby)

```python
# Average Highest Attact pokemon type
df.groupby('Type 1').mean().sort_values('Attack',ascending= False)
```

```python
# Average Highest Defense pokemon type
df.groupby('Type 1').mean().sort_values('Defense',ascending= False)
```

```python
# Average Highest Speed pokemon type
df.groupby('Type 1').mean().sort_values('Speed',ascending= False)
```

```python
# Number of pokemons of each type 1
df.groupby('Type 1').count()['Name']
```

```python
df.groupby(['Type 1','Type 2']).count()['Name']
```
